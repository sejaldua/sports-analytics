## Shot Chart Analysis

```python
schedule = get_schedule(2020, playoffs=False)
schedule = schedule[(schedule['HOME'] == 'Portland Trail Blazers') | (schedule['VISITOR'] == 'Portland Trail Blazers')]
schedule
dates = schedule['DATE'].to_list()
visitor_list = schedule['VISITOR'].to_list()
home_list = schedule['HOME'].to_list()
```

```python
shot_df = pd.DataFrame(columns=['x', 'y', 'QUARTER', 'TIME_REMAINING', 'PLAYER', 'MAKE_MISS', 'VALUE', 'DISTANCE'])
for i, date in enumerate(dates):
    shot_df = pd.concat([shot_df, get_shot_chart(date, TEAM_ABBREVS[home_list[i].upper()], TEAM_ABBREVS[visitor_list[i].upper()])['POR']], ignore_index=True)
shot_df
```

```python
s = shot_df.copy()
s['x_pos'] = s['x'].apply(lambda x: float(x.strip(' ft')))
s['y_pos'] = s['y'].apply(lambda y: float(y.strip(' ft')))
# sns.heatmap(data=s.pivot("x_pos", "y_pos", "hue"))
x = s['x_pos'].to_list()
y = s['y_pos'].to_list()
plt.hist2d(x, y, bins=22, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```
